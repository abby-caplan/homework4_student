## Part 1: Loading data and cleaning (20 points)
1) Store the name of your file in `filepath` and open it with the `fileobj` variable. Use the `readline()` function and a for loop, find out how many lines of header are in the "Pakistan_sea_level_psmsl.txt" file and store this number in the `header_lines` variable. Read the information in the header. In this data file, missing sea level data points are filled with a placeholder "masking" value. (5 points)
>1) What is the mask value for missing data in this dataset? Answer this question in the markdown cell below the solution cell.

2) Use numpy `genfromtext()` to read each column of your text file into a Numpy array as floats and store the array in the `sea_level` variable.(_HINT: Remember that indexing starts with 0 in Python!_). Store the columns as individual 1-D arrays (excluding "number of missing days") in the provided answer variables, `year_month`, `depth_mm`, `data_flag`. (5 points)
>1) What is the delimiter type for this text file? Answer this question in the markdown cell below the solution cell.

3) Using the numpy `where()` function and its first positional argument ([Documentation](https://numpy.org/doc/stable/reference/generated/numpy.where.html)), find the indices of the masked depth values that represent missing values in your Numpy array and store the resulting array in `mask_inds`. Using `mask_inds`, replace the masking values in `depth_mm` with `np.nan` values. (5 points)

4) PSMSL also flags data that is likely erroneous. In this data, a data quality flag value of 11 represents bad data. We will also replace these suspicious values with `np.nan` values to ignore them. (5 points)

  >a) Find the indices of bad data quality values in `data_flag` by numpy `where()` and store in `bad_inds`.

  >b) Save the corresponding "year" and "depth_mm" values of bad data in a separate arrays, `bad_years` and `bad_depths` respectively. 
  
  >c) Replace bad depth values in `depth_mm` with `np.nan` values.

````python
# import the necessary packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import interpolate 
from scipy import stats
print("Part 1")
# your code

# answer variables 
filepath = 'Pakistan_sea_level_psmsl.txt' #file name
fileobj = open(filepath) # open file

#prints line number and one line of file 40 times
for x in range(1, 40):
    print(x)
    print(fileobj.readline())

header_lines = 22 # counted from previous for loop

print("Part 2")
# your code


# answer variables
sea_level = np.genfromtxt(filepath, delimiter = ';', skip_header = header_lines) # generates numpy array from data with 22 header lines and semicolon delimiter
year_month = sea_level[:,0] # selects first column (year)
depth_mm = sea_level[:,1] # selects second column (depth)
data_flag = sea_level[:,3] # selects last column (flag)

#display


print("Part 3")
# your code


# answer variables
mask_inds = np.where(depth_mm == -99999) # finds indices where depth is the mask value
# replaces depth at mask indices with np.nan
for x in mask_inds:
    depth_mm[x] = np.nan

print("Part 4")
# your code

# answer variables
bad_inds = np.where(data_flag == 11) # finds indices where flag = 11

badyears = year_month[bad_inds] # finds months corresponding to bad_inds
baddepths = depth_mm[bad_inds] # ^ for depths

# replaces depth at bad indices with np.nan
for x in bad_inds:
    depth_mm[x] = np.nan
````

The mask value is -99999. The delimiter for this data is a semicolon.

2) The `rolling()` function does not deal with nan values, which represents our missing and suspicious data. This results in creating larger data gaps in the calculated rolling mean depth values compared to the original data. Here, we will _interpolate and extrapolate_ the rolling mean values to resolve the data gaps. (10 points)

>a) Apply the `interpolate()` function to the `df` to the df["yearly_height"] column. Set the argument `limit_direction = "both"` to extrapolate the data at the ends. Store your interpolated depth values in a new column called "Depth_interp_roll" in `df`.

>c) Do you think this interpolation describes the trends in missing depth data more or less accurately than your first interpolation attempt? Why or why not? Store your answer in the markdown cell below the solution cell. 

````python
print('Part 1: ')
# your code
df['yearly_height'] = df['Depth'].rolling(12).mean() # 12 month (1 year) roling mean of 'Depth'


print('Part 2: ')
# your code 
df['Depth_interp_roll'] = df['yearly_height'].interpolate(limit_direction = 'both') # interpolated 'yearly_heights'

# display
display(df)

# your code for plotting
fig, ax = plt.subplots(nrows = 2, figsize = (10, 6), sharex = True, sharey = True)
ax[0].plot(year_month, depth_mm, '-^', label = 'Original')
ax[0].plot(badyears, baddepths, '-^', color = 'red', label = 'Flagged')
ax[0].plot(df['Year'], df['yearly_height'], '--', linewidth = 3, color = 'purple', label = 'Yearly Rolling Mean') # Rolling mean
ax[0].set_title('Pakistan Sea Level')
ax[0].set_ylabel('Sea Level (mm)')
ax[0].legend()

ax[1].plot(df['Year'], df['Depth_interp'], '-^', color = 'darkgreen', label = 'Interpolated')
ax[1].plot(df['Year'], df['Depth_interp_roll'], '--', color = 'navy', linewidth = 3, label = 'Interpolated Yearly Rolling Mean') # adding interpolated rolling mean
ax[1].set_ylabel('Sea Level (mm)')
ax[1].set_xlabel('Time')
ax[1].legend()
````

I think this interpolation is much more accurate because it interpolates based on a mean of the chunk of data before and after. This removes the steep trends that probably don't actually exist.

````python
print("Part 1: ")
# your code to calculate the linear regression
slope, intercept, rvalue, pvalue, stderr = stats.linregress(df['Year'], df['Depth_interp']) # linear regression of interpolated depth and year

# your code to calculate the linear regression fit
df['full_regression'] = (df['Year']*slope) + intercept # y = mx + b for each time

print(slope)


print("Part 2: ")
# your code
recent_df = df.loc[df['Year'] > 2007] # subsetting to only recent data

# your code to calculate the linear regression for recent data
slope_new, intercept_new, rvalue_new, pvalue_new, stderr_new = stats.linregress(recent_df['Year'], recent_df['Depth_interp']) # linear regression of depths after 2007

print(slope_new)
# your code to calculate the linear regression fit for recent data
df['recent_regression'] = (df['Year']*slope_new) + intercept_new # y = mx + b for each time

# your code to plot
fig, ax = plt.subplots(nrows = 2, figsize = (10, 10), sharex = True, sharey = True)
ax[0].plot(year_month, depth_mm, '-^', label = 'Original')
ax[0].plot(badyears, baddepths, '-^', color = 'red', label = 'Flagged')
ax[0].plot(df['Year'], df['yearly_height'], '--', linewidth = 3, color = 'purple', label = 'Yearly Rolling Mean')
ax[0].set_title('Pakistan Sea Level')
ax[0].set_ylabel('Sea Level (mm)')
ax[0].legend()

ax[1].plot(df['Year'], df['Depth_interp'], '-^', color = 'darkgreen', label = 'Interpolated')
ax[1].plot(df['Year'], df['Depth_interp_roll'], '--', color = 'navy', linewidth = 3, label = 'Interpolated Yearly Rolling Mean')
# Regression plots
ax[1].plot(df['Year'], df['full_regression'], '-.', color = 'pink', linewidth = 3, label = 'Full Regression (rate of rise = 1.96 mm/year')
ax[1].plot(df['Year'], df['recent_regression'], '-.', color = 'cyan', linewidth = 3, label = 'Recent Regression (rate of rise = 3.57 mm/year')

ax[1].set_ylabel('Sea Level (mm)')
ax[1].set_xlabel('Time')
ax[1].legend()
````

I think that the rate of sea level rise in Pakistan is increasing because the slope calculated from the recent data (3.57 mm/y) is greater than the slope from all of the data combined (1.96 mm/7).
